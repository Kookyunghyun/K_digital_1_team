```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline  

import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도

#한글 폰트 설정
plt.rc('font', family='NanumBarunGothic')
```

```python
#데이터 불러오기

road = pd.read_csv('/content/drive/MyDrive/2023/k-디지털/지역별인구수.csv',  encoding = "euc-kr", engine='python')
road.head()
```

```python
#히스토그램 그리기 - 총 인구밀도


#특정 열 값들을 리스트로 변환
area = road['시도명']
human = road['인구수']

index = np.arange(len(human))
bar_width = 0.5
opacity = 0.4
error_config = {'ecolor': '0.3'}
rects1 = plt.bar(index, human, bar_width, alpha=opacity, color='b',
                 yerr=None,error_kw=error_config,label='인구 밀도')

plt.xlabel('구 별')
plt.ylabel('인구수')
plt.title('인구 밀도')
plt.xticks(index, area)
plt.legend()
plt.tight_layout() 
plt.rc('font', size=6)

plt.show()

#png 파일로 저장
plt.savefig('Human.png')
```

```python
#히스토그램그리기 - 성별인구비교

#특정 열 값들 리스트로 변환
area = road['시도명']
boy = road['남성']
girl = road['여성']

index = np.arange(len(boy)) 
bar_width = 0.35
opacity = 0.4
error_config = {'ecolor': '0.3'}
rects1 = plt.bar(index, boy, bar_width, alpha=opacity, color='b', 
                 yerr=None, error_kw=error_config,label='남성')
rects2 = plt.bar(index + bar_width, girl, bar_width, alpha=opacity, color='r',
                 yerr=None, error_kw=error_config, label='여성')

plt.xlabel('구 별')
plt.ylabel('인구 수')
plt.title('지역별 성별 인구 수 비교')
plt.xticks((index + bar_width)-(bar_width/2), area)
plt.legend()
plt.tight_layout()
plt.rc('font', size=6)

plt.show()
#png 파일로 저장
plt.savefig('Human_gender.png')
```
