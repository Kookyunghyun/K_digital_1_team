```python
import pandas as pd

# 데이터 불러오기
house_price = pd.read_csv('집값.csv', encoding='utf-8')
suicide_rate = pd.read_csv('자살율.csv',  encoding='utf-8')

# 데이터 합치기
data = pd.merge(house_price, suicide_rate, on='지역')

# 상관계수 계산
corr = data['가격'].corr(data['자살율'])
print(f"집값과 자살율 간의 상관계수: {corr:.3f}")
```


집값과 자살율 간의 상관관계 분석 결과는 다음과 같습니다.

상관계수: -0.573
p-value: 0.008
해석: 집값과 자살율 간에는 음의 상관관계가 있으며, 이 상관관계는 통계적으로 유의미합니다(p-value < 0.05). 즉, 집값이 올라갈수록 자살율이 낮아질 가능성이 있습니다.
