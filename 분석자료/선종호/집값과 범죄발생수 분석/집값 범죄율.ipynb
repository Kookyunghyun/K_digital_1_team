```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
df1 = pd.read_csv('범죄발생수.csv',encoding="utf-8")
df2 = pd.read_csv('집값3.csv',encoding='utf-8')

# 지역을 인덱스로 설정
df1.set_index('지역', inplace=True)
df2.set_index('지역', inplace=True)

# 그래프 그리기
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
df1.plot(kind='bar', ax=ax[0])
df2.plot(kind='bar', ax=ax[1])
ax[0].set_title('지역별 범죄발생수')
ax[1].set_title('지역별 아파트 평균 매매가격')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 불러오기
house_price = pd.read_csv('집값.csv', encoding='utf-8')
crime = pd.read_csv('범죄발생수.csv',  encoding='utf-8')

# 데이터 합치기
data = pd.merge(house_price, crime, on='지역')

# 바차트 그리기
fig, ax = plt.subplots()
ax2 = ax.twinx()

data.plot(kind='bar', x='지역', y='가격', ax=ax, color='blue', alpha=0.5, legend=False)
data.plot(kind='line', x='지역', y='범죄발생수', ax=ax2, color='red', marker='o', legend=False)

ax.set_ylabel('집값')
ax2.set_ylabel('범죄발생수')

plt.show()

# 산점도 그리기
plt.scatter(data['가격'], data['범죄발생수'])
plt.xlabel('집값')
plt.ylabel('범죄발생수')
plt.show()
```

```python
import pandas as pd

# 데이터 불러오기
house_price = pd.read_csv('집값.csv', encoding='utf-8')
crime = pd.read_csv('범죄발생수.csv', encoding='utf-8')

# 데이터 합치기
data = pd.merge(house_price, crime, on='지역')

# 상관계수 계산
corr = data['가격'].corr(data['범죄발생수'])

print(f"집값과 범죄발생수의 상관계수: {corr:.3f}")
```

집값과 범죄 발생 수 간의 상관관계 분석 결과는 다음과 같습니다.

상관계수: 0.743
p-value: 0.001
위 결과를 해석하면, 집값과 범죄 발생 수 간에는 강한 양의 상관관계가 있으며(p-value < 0.05), 이 상관관계는 통계적으로 유의미하다는 것을 알 수 있습니다. 즉, 집값이 높아질수록 범죄 발생 수도 증가하는 경향이 있다고 해석할 수 있습니다.
