```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 한글 폰트 설정
mpl.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 불러오기
crime_df = pd.read_csv('범죄발생수2.csv', encoding='utf-8')
house_price_df = pd.read_csv('집값.csv', encoding='utf-8')

# 데이터 합치기
data = pd.merge(house_price_df, crime_df, on='지역')

# 집값을 내림차순으로 정렬
data = data.sort_values(by='가격', ascending=False)


# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(15, 8))
ax2 = ax1.twinx()

x = data['지역']
width = 0.4

# 집값 바차트
ax1.bar(x, data['가격'], width=width, color='blue', alpha=0.5, label='집값')
ax1.set_ylabel('집값', color='blue')

# 강력범죄 선차트
ax2.plot(x, data['강력범죄'], color='red', marker='o', linewidth=2, markersize=6, label='강력범죄')
ax2.set_ylabel('강력범죄', color='red')

# 절도범죄 선차트
ax2.plot(x, data['절도범죄'], color='green', marker='o', linewidth=2, markersize=6, label='절도범죄')

# 폭력범죄 선차트
ax2.plot(x, data['폭력범죄'], color='orange', marker='o', linewidth=2, markersize=6, label='폭력범죄')

ax2.set_ylabel('범죄 발생 건수', color='black')
ax2.grid(False)

ax1.set_xlabel('지역')
ax1.set_title('집값과 범죄 발생 비교')

# 범례 추가
ax1.legend(loc='upper left', bbox_to_anchor=(0, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(1, 1))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 데이터 불러오기
crime_df = pd.read_csv('범죄발생수2.csv', encoding='utf-8')
house_price_df = pd.read_csv('집값.csv', encoding='utf-8')

# 데이터 합치기
merged_df = pd.merge(house_price_df, crime_df, on='지역')

# 상관관계 분석
correlation = merged_df[['가격', '강력범죄', '절도범죄', '폭력범죄']].corr()

# 상관계수 출력
print(correlation)
```

--집값과 강력범죄: 상관계수 0.663639 (양의 상관관계)
--집값과 절도범죄: 상관계수 0.759329 (양의 상관관계)
--집값과 폭력범죄: 상관계수 0.722704 (양의 상관관계)

--집값과 강력범죄, 절도범죄, 폭력범죄 간에 양의 상관관계가 있음을 확인하였습니다.
집값이 상승할수록 강력범죄, 절도범죄, 폭력범죄 발생 건수도 증가하는 경향을 보입니다.
상관계수는 집값과 각 범죄 유형 사이의 상관관계의 강도를 나타내며, 양의 상관관계일수록 상승하는 경향을 보입니다.
이러한 결과는 집값과 범죄 사이에 어느 정도의 연관성이 있음을 시사하지만, 인과관계를 나타내지는 않습니다.
집값의 상승이 범죄 발생을 직접적으로 유발한다고 일반화할 수는 없으며, 추가적인 분석과 고려가 필요합니다.
